```python
from collections import Counter
def get_seq(x, M):
    index_for_length_m = M - 1
    for v in [l for l in x if len(l) >= M]:
        for i in [i for i, v in enumerate(v[index_for_length_m:], start=index_for_length_m)]:
            # convert to str to be hashable
            yield str(v[i - index_for_length_m : i + 1])

def process_chunk(x, M, N):
    return Counter(get_seq(x, M)).most_common(N)
```

```python
def getUniqueItems(d):
    result = {}
    for key,value in d.items():
        if value not in result.values():
            result[key] = value
    return(result)
    
def combinations_rules(rule):
    tag_list = sorted(rules_with_tags[rule], key=itemgetter('start'))
    list_starts = []
    for tag in tag_list:
        list_starts.append([tag['start'],tag['end']])
    merged = [list_starts[0]]
    for current in list_starts:
        previous = merged[-1]
        if current[0] <= previous[1]:
            previous[1] = max(previous[1], current[1])
        else:
            merged.append(current)
    combination_dict = {}
    for k in merged:
        combination_dict[tuple(k)] = []
    for tag in tag_list:
        for interval in combination_dict:
            if tag['start']>=interval[0] and tag['end']<=interval[1]:
                combination_dict[interval].append(tag)
    print(rule)
#     return combination_dict
    #generate tag sequences.
    all_optional_sequences = []
    for interval in combination_dict:
        tags = []
        for r in combination_dict[interval]:
            tags.append(r['label'])
        all_optional_sequences.append(tags)
    combinations = list(itertools.product(*all_optional_sequences))
    return combinations
```
