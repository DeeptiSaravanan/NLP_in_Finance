```python
import pandas as pd
import json
```

```python
with open('db_dump.json','r') as f:
    json_obj = json.load(f)
```

```python
df = pd.DataFrame.from_dict(json_obj['_default'],orient='index')
```

```python
def generate_file_name(row):
    if type(row.file_url) == str:
        return row.file_url.split('/')[-1]
    else:
        return 'Nan'
```

```python
df['file_name'] = df.apply(generate_file_name, axis = 1) 
```
